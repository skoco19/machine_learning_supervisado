```python
import pandas as pd
import numpy as np
```

```python
# PREPROCESAMIENTO BASICO
# 1. Como tratar los nulls.
# 2. Como transformar las variables categóricas o strings
# 3. Es necesario transformar la distribución de algunas variables?.
# 4. como gestiono los outliers.
# 5. Necesito estandarizar los valores. 
```

```python
# 1. NULLs:
# 1.1 No hacer nada. (para HistGradientBoost, XGBoost, LightGBM)
# 1.2 Eliminar filas o columnas con Null. (cuando hay muy pocas filas o muchas columnas)
# 1.3 Imputación, basada en más frecuente, media, mediana, k-nn.
# 1.4 Es darle un valor diferente al resto de datos. Si es catégorico : NO disponible, si
# numérico valor alejado.

df_train = pd.read_csv("data/preprocessed/train_data.csv")
```

```python
df_train = df_train.replace({9:np.nan})
```

```python
df_train.isna().sum()
```

```python
df_w_bool = df_train.copy()
df_w_bool[df_w_bool.columns[5:]] = df_w_bool[df_w_bool.columns[5:]].replace({0:'NO', 1:'SI', np.nan:'NO_DISPONIBLE'})
```

```python
df_w_bool
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='median')
df_w_bool['No. of Bedrooms'] = imputer.fit_transform(df_w_bool[['No. of Bedrooms']])
```

```python
df_w_bool.isna().sum()
```

```python
# 2. VARIABLES CATEGORICAS (EXCEPTO TEXTOS O VARIABLES CON MUCHAS CATEGORIAS)
# 2.1. OrdinalEncoding|LabelEncoding (0,1,2, etc..).
#       - Árboles de decisión.
#       - Hay un orden implícito.
# 2.2 OneHotEncoding. Más usada, y recomendada por defecto.
# 2.3 TargetEncoding.

# procesar todas las variables de baja cardinalidad con one hot encoding.

from sklearn.preprocessing import OneHotEncoder
```

```python
#instanciar el objeto
ohEnconder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
```

```python
#extraer datos 

cat_columns = ["city"] + df_w_bool.columns[5:].to_list()
cat_columns

# hacemos fit con datos cat.
ohEnconder.fit(df_w_bool[cat_columns])
```

```python
cat_oh_data = ohEnconder.transform(df_w_bool[cat_columns])
```

```python
cat_oh_data
```

```python
df_cat_oh_data = pd.DataFrame(data=cat_oh_data, columns=ohEnconder.get_feature_names_out(), index=df_train.index)
```

```python
df_cat_oh_data.head()
```

```python
df_train['Location'].nunique()
```

```python
# procesar location.
from skrub import GapEncoder

gapEncoder = GapEncoder(n_components=50)
```

```python
gapEncoder.fit(df_train['Location'])
```

```python
gapEncoder.get_feature_names_out()
```

```python
cat_location_data = gapEncoder.transform(df_train['Location'])
cat_location_data
```

```python
from sklearn.preprocessing import QuantileTransformer
```

```python
price_Transformer, area_Transformer, beds_Transformer = QuantileTransformer(), QuantileTransformer(), QuantileTransformer()
```

```python
price_Transformer.fit(df_train[['Price']])
area_Transformer.fit(df_train[['Area']])
beds_Transformer.fit(df_train[['No. of Bedrooms']])
```

```python
price_normal = price_Transformer.transform(df_train[['Price']])
area_normal = area_Transformer.transform(df_train[['Area']])
beds_normal = beds_Transformer.transform(df_train[['No. of Bedrooms']])
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler_price, scaler_area, scalar_beds = StandardScaler(), StandardScaler(), StandardScaler()
```

```python
scaler_price.fit(price_normal)
scaler_area.fit(area_normal)
scalar_beds.fit(beds_normal)

price_normal_scaled = scaler_price.transform(price_normal)
area_normal_scaled = scaler_area.transform(area_normal)
beds_normal_scaled = scalar_beds.transform(beds_normal)
```

```python
area_normal_scaled.flatten()
```

```python
df_normal_scaled = pd.DataFrame(data={'price_normal_scaled':price_normal_scaled.flatten(), 
                                      'area_normal_scaled':area_normal_scaled.flatten(), 
                                      'beds_normal_scaled':beds_normal_scaled.flatten()}, 
                                      index=df_train.index)
```

```python
df_normal_scaled.head()
```

```python
print(df_cat_oh_data.shape)
print(cat_location_data.shape)
print(df_normal_scaled.shape)
```

```python
# Enriquecer nuestras variables con nuevas features.
# 1.- Añadir features espécificas según negocio.
# 2.- Si tenemos datos temporales, extraer, mes, año, estación, fin de semana, etc..
# 3.- si tuvieramos localización GPS, extraer, ciudad, barrio, etc..
# 4.- Ver interacciones cruzadas.
# 5.- Crear clusters y añadirlos como variables.
# 6.- Crear PCA y añadirlo como variables,
# 7.- etc...

# En este ejemplo de momento solo vamos a añadir interacciones cruzadas.

from sklearn.preprocessing import PolynomialFeatures
```

```python
polyfeatures = PolynomialFeatures(degree=2, interaction_only=True, include_bias=False)

features_to_cross = pd.concat([df_cat_oh_data, df_normal_scaled[['area_normal_scaled', 'beds_normal_scaled']]], axis=1)
```

```python
features_to_cross.shape
```

```python
polyfeatures.fit(features_to_cross)
```

```python
crossed_features = polyfeatures.transform(features_to_cross)
```

```python
crossed_features.shape
```

```python
df_crossed_features = pd.DataFrame(data=crossed_features, columns=polyfeatures.get_feature_names_out(), index=df_train.index)
```

```python
df_crossed_features
```

```python
# generar Matriz X y vector y

X_EXPANDED = pd.concat([df_normal_scaled[['area_normal_scaled', 'beds_normal_scaled']],
                        df_cat_oh_data,
                        cat_location_data,
                        df_crossed_features], axis=1)
y = df_normal_scaled['price_normal_scaled']
```

```python
X_EXPANDED.shape
```

```python
y.shape
```

```python
# guardamos los dataframes en modo pickle
DIR = "data/postproccesed/"
df_normal_scaled.to_pickle(DIR + 'df_normal_scaled.pkl')
df_cat_oh_data.to_pickle(DIR + 'df_cat_oh_data.pkl')
cat_location_data.to_pickle(DIR + 'cat_location_data.pkl')
df_crossed_features.to_pickle(DIR + 'df_crossed_features.pkl')
# guardamos también X e Y
X_EXPANDED.to_pickle(DIR +'X_EXPANDED.pkl')
y.to_pickle(DIR +'y.pkl')
```

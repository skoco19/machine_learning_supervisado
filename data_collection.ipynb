```python
import os
```

```python
dirs_files = [os.path.join(ROUTE, f) for f in os.listdir(ROUTE) if f.endswith('.csv')]
```

```python
dirs_files
```

```python
dfs = []
for file in dirs_files:
    df = pd.read_csv(file)
    city_name_1 = file.split('/')[-1]
    city_name = city_name_1.split('.csv')[0]
    df.insert(1,"city",city_name)
    dfs.append(df)
```

```python
[df.shape for df in dfs]
```

```python
df_combined = pd.concat(dfs, ignore_index=True)
```

```python
df_combined.shape
```

```python
df_combined = df_combined.sample(frac=1)
```

```python
df_combined.head()
```

```python
df_combined.to_csv("data/preprocessed/combine_data.csv", index=False)
```

```python
from sklearn.model_selection import train_test_split
train_df, test_df = train_test_split(df_combined,test_size=0.3, random_state=42)
```
